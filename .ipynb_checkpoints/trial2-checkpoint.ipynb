### Question 8b,c,d,e)

```python
model = InceptionV3(include_top=True, weights='imagenet', input_tensor=None, input_shape=(299,299,3), pooling=None, classes=1000)
```

```python
def image_preprocess(path):
    img_path = path
    img = image.load_img(img_path, target_size=(299, 299))
    plt.imshow(img)
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    return x
```

```python
snake = image_preprocess('data\snake.jpg')
preds_snake = model.predict(snake)
print('Predicted:', decode_predictions(preds_snake, top=5)[0])
```

```python
koala = image_preprocess('data\koala.jpg')
preds_koala = model.predict(koala)
print('Predicted:', decode_predictions(preds_koala, top=5)[0])
```
